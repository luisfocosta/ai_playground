Next we need to prepare to load a range of quantized language models, optimized for memory efficiency with 4-bit quantization.
##Reduced Memory Usage:

Quantizing a model to 4-bit precision significantly reduces the amount of memory required to store the model. This is particularly beneficial for deploying large models on devices with limited memory capacity.

# Faster Inference

Lower precision models can perform computations faster because they require fewer bits to represent each parameter. This leads to quicker inference times, making the models more efficient for real-time applications.

# Cost Efficiency:
Reduced memory and faster inference translate to lower computational costs. This is especially important for large-scale deployments where resource efficiency can lead to substantial cost savings.

# Maintained Performance
With advanced quantization techniques, 4-bit models can maintain a high level of performance and accuracy. For larger models (over ~70B parameters), the performance impact of quantization is minimal, making 4-bit a good compromise between size and performance.

# Energy Efficiency
Lower precision computations consume less power, making 4-bit models more energy-efficient. This is crucial for sustainable AI practices and deploying models in energy-constrained environments.

```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # Choose any! Llama 3 is up to 8k
dtype = None
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.

fourbit_models = [
    "unsloth/mistral-7b-bnb-4bit",
    "unsloth/mistral-7b-instruct-v0.2-bnb-4bit",
    "unsloth/llama-2-7b-bnb-4bit",
    "unsloth/gemma-7b-bnb-4bit",
    "unsloth/gemma-7b-it-bnb-4bit",
    "unsloth/gemma-2b-bnb-4bit",
    "unsloth/gemma-2b-it-bnb-4bit",
    "unsloth/llama-3-8b-bnb-4bit",
]

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/llama-3-8b-bnb-4bit", # change the model name here
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    # token = "hf_...", # use one if using gated models like meta-llama/Llama-2-7b-hf
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # Choose any number > 0 ! Suggested 8, 16, 32, 64, 128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 3407,
    use_rslora = False,
    loftq_config = None,
)
```

Integrate LoRA adapters into our model, which allows to efficiently update just a fraction of the model's parameters, enhancing training speed and reducing computational load.
# Why LoRa?
## Efficiency
LoRA adapters significantly reduce the computational resources required for fine-tuning. Instead of updating all the model's parameters, LoRA fine-tunes only a small subset of parameters, which are represented as low-rank matrices.
## Memory Usage
By focusing on a smaller number of parameters, LoRA adapters drastically cut down on memory usage. This makes it feasible to fine-tune large models on hardware with limited memory capacity.
## Cost-Effectiveness
Fine-tuning with LoRA adapters is more cost-effective because it requires fewer computational resources and less time. This is particularly beneficial for startups and small businesses that may not have access to high-end GPUs.
## Modularity
LoRA adapters are modular, allowing you to easily switch between different tasks without retraining the entire model. This flexibility makes it easier to adapt models to various applications.
## Preservation of Base Model
LoRA allows the base model to remain unchanged while adding task-specific adaptations. This means you can maintain the general capabilities of the pre-trained model while fine-tuning it for specific tasks.
## Scalability
LoRA adapters enable scalable fine-tuning of large models, making it possible to handle specialized requirements without running a full fine-tuning process.

# Data Preparation
We now use the Alpaca dataset from [yahma](https://www.google.com/url?q=https%3A%2F%2Fhuggingface.co%2Fdatasets%2Fyahma%2Falpaca-cleaned/), which is a filtered version of 52K of the [original Alpaca dataset](https://www.google.com/url?q=https%3A%2F%2Fcrfm.stanford.edu%2F2023%2F03%2F13%2Falpaca.html).

Then, we define a system prompt that formats tasks into instructions, inputs, and responses, and apply it to a dataset to prepare our inputs and outputs for the model, with an EOS token to signal completion.

```python
# this is basically the system prompt
alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.
### Instruction:
{}
### Input:
{}
### Response:
{}"""
EOS_TOKEN = tokenizer.eos_token # This marks the end of the prompt.
def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    inputs       = examples["input"]
    outputs      = examples["output"]
    texts = []
    for instruction, input, output in zip(instructions, inputs, outputs):
        text = alpaca_prompt.format(instruction, input, output) + EOS_TOKEN # without this token generation goes on forever!
        texts.append(text)
    return { "text" : texts, }
pass

from datasets import load_dataset
dataset = load_dataset("yahma/alpaca-cleaned", split = "train") #this splits up the dataset automatically with an existing dataset tag
dataset = dataset.map(formatting_prompts_func, batched = True,)
```

# Train the model


*   We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`.

*   At this stage, we're configuring our model's training setup, where we define things like batch size and learning rate, to teach our model effectively with the data we have prepared.

```python
from trl import SFTTrainer
from transformers import TrainingArguments

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    dataset_num_proc = 2,
    packing = False, # Can make training 5x faster for short sequences.
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 60,
        num_train_epochs=4,
        learning_rate = 2e-4,
        fp16 = not torch.cuda.is_bf16_supported(),
        bf16 = torch.cuda.is_bf16_supported(),
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
    ),
)
```

```python
# We're now kicking off the actual training of our model, which will spit out some statistics showing us how well it learns
trainer_stats = trainer.train()
```

```python
#@title Show final memory and time stats
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round(used_memory         /max_memory*100, 3)
lora_percentage = round(used_memory_for_lora/max_memory*100, 3)
print(f"{trainer_stats.metrics['train_runtime']} seconds used for training.")
print(f"{round(trainer_stats.metrics['train_runtime']/60, 2)} minutes used for training.")
print(f"Peak reserved memory = {used_memory} GB.")
print(f"Peak reserved memory for training = {used_memory_for_lora} GB.")
print(f"Peak reserved memory % of max memory = {used_percentage} %.")
print(f"Peak reserved memory for training % of max memory = {lora_percentage} %.")
```

# Inference
Let's run the model! You can change the instruction and input - leave the output blank!

```python
FastLanguageModel.for_inference(model)
inputs = tokenizer(
[
    alpaca_prompt.format(
        "List the prime numbers contained within the range.", # instruction
        "1-50", # input
        "", # output - leave this blank for generation!
    )
], return_tensors = "pt").to("cuda")

outputs = model.generate(**inputs, max_new_tokens = 128, use_cache = True)
tokenizer.batch_decode(outputs)
```

```python
FastLanguageModel.for_inference(model)
inputs = tokenizer(
[
    alpaca_prompt.format(
        "Convert these binary numbers to decimal.", # instruction
        "1010, 1101, 1111", # input
        "", # output - leave this blank for generation!
    )
], return_tensors = "pt").to("cuda")

from transformers import TextStreamer
text_streamer = TextStreamer(tokenizer)
_ = model.generate(**inputs, streamer = text_streamer, max_new_tokens = 128)
```
